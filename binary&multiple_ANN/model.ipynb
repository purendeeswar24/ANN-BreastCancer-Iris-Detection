```python
def out(col):
    q1 , q3 = np.percentile(col,[25,75])
    iqr = q3-q1
    ll = q1-1.5*iqr
    ul = q3+1.5*iqr
    return ll , ul 

for i in df.select_dtypes(include="number").columns:
    ll , ul = out(df[i])
    df[i] = np.where(df[i]<ll , ll , df[i])
    df[i] = np.where(df[i]>ul , ul , df[i])
```

```python
for i in df.select_dtypes(include="number").columns:
    sns.boxplot(df[i])
    plt.show()
```

```python
d={"M":0,"B":1}
df["diagnosis"]=df["diagnosis"].map(d)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.22, random_state=45)

# One-hot encode labels for multi-class classification
y_train = to_categorical(y_train, num_classes=3)  # Assume 3 classes
y_test = to_categorical(y_test, num_classes=3)

# Model setup
model = Sequential()
model.add(Dense(150, activation="relu", input_dim=x_train.shape[1]))
model.add(Dense(75, activation="relu"))
model.add(Dense(15, activation="relu"))
model.add(Dense(3, activation="sigmoid"))  # For multi-class classification

# Compile the model
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

# Train the model
history = model.fit(x_train, y_train, epochs=5, batch_size=10, validation_split=0.2)
```
